```python
# Plot dos gráficos
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Pandas
import pandas as pd
```

```python
import seaborn as sns
from matplotlib import colors
import scipy.cluster.hierarchy as sch # Dendograma
```

```python
# Avaliação de desempenho
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
# Algortmos de Agrupamentos
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
```

```python
url_bd = 'https://raw.githubusercontent.com/reisanar/datasets/master/moons.csv'
```

```python
moons = pd.read_csv(url_bd)
```

```python
moons.head()
```

```python
# Grafico de DISPERSÃO
plt.scatter(x=moons['X'], y=moons['Y'])
```

```python
# Modelos de agrupamento
rs = 42
km = KMeans(n_clusters=2, random_state=rs)          #KM k=4
gm = GaussianMixture(n_components=4, random_state=rs)  # Gussian Mixture k=4
db = DBSCAN(eps=0.3)
```

```python
# Aplica o algoritmo e armazena o cluster de cada dado
km_c = km.fit_predict(moons) #
gm_c = gm.fit_predict(moons) # Permite mistura entre grupos (bem parecido com k menos)
db_c = db.fit_predict(moons) # Permite formatos não globulares
```

```python
km_c
```

```python
gm_c
```

```python
db_c
```

```python
fig = plt.figure(figsize=(12, 8), dpi=80)

plt.subplot(2,2,1)
plt.scatter(x=moons['X'], y=moons['Y'])
plt.title('Original')

plt.subplot(2,2,2)
plt.scatter(x=moons['X'], y=moons['Y'], c=km_c)
plt.title('KMeans')

plt.subplot(2,2,3)
plt.scatter(x=moons['X'], y=moons['Y'], c=gm_c)
plt.title('GaussianMixture')

plt.subplot(2,2,4)
plt.scatter(x=moons['X'], y=moons['Y'], c=db_c)
plt.title('DBSCAN')

plt.subplots_adjust(hspace = 0.35)
```
